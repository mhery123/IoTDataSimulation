```python
from web3 import Web3
import json

# Connect to Ganache
ganache_url = "http://127.0.0.1:7545"  # Change port if yours is different
web3 = Web3(Web3.HTTPProvider(ganache_url))

# Check connection
if web3.is_connected():
    print("Connected to Ganache successfully!")
else:
    print("Connection failed. Check if Ganache is running.")

# Your contract details (from Week 3)
contract_address = "0x1853601d8894631c13495947A26E26bA0f7B7de5"  # Paste your deployed contract address
deployer_address = "0xA6B5f8679fa38db716c993fE425299cAAC690698"  # Paste your owner/deployer address

# ABI (paste your full ABI from Week 3 here)
abi = json.loads('[{"anonymous":false,"inputs":[{"indexed":false,"internalType":"string","name":"patientID","type":"string"},{"indexed":false,"internalType":"string","name":"dataType","type":"string"},{"indexed":false,"internalType":"string","name":"value","type":"string"},{"indexed":false,"internalType":"uint256","name":"timestamp","type":"uint256"}],"name":"DataStored","type":"event"},{"inputs":[{"internalType":"string","name":"patientID","type":"string"},{"internalType":"string","name":"dataType","type":"string"},{"internalType":"string","name":"value","type":"string"}],"name":"storeData","outputs":[],"stateMutability":"nonpayable","type":"function"},{"inputs":[{"internalType":"uint256","name":"index","type":"uint256"}],"name":"getRecord","outputs":[{"internalType":"string","name":"","type":"string"},{"internalType":"string","name":"","type":"string"},{"internalType":"string","name":"","type":"string"},{"internalType":"uint256","name":"","type":"uint256"}],"stateMutability":"view","type":"function"},{"inputs":[],"name":"getTotalRecords","outputs":[{"internalType":"uint256","name":"","type":"uint256"}],"stateMutability":"view","type":"function"},{"inputs":[{"internalType":"uint256","name":"","type":"uint256"}],"name":"records","outputs":[{"internalType":"string","name":"patientID","type":"string"},{"internalType":"string","name":"dataType","type":"string"},{"internalType":"string","name":"value","type":"string"},{"internalType":"uint256","name":"timestamp","type":"uint256"}],"stateMutability":"view","type":"function"}]')
# Load the contract
contract = web3.eth.contract(address=contract_address, abi=abi)
print("Smart contract recognized!")

# Store IoT data on the blockchain
def store_health_data(patient_id, data_type, value):
    tx = contract.functions.storeData(patient_id, data_type, value).transact({
        "from": deployer_address
    })
    receipt = web3.eth.wait_for_transaction_receipt(tx)
    print(f"Data stored! Transaction hash: {receipt.transactionHash.hex()}")

# Retrieve a record
def get_health_data(index):
    record = contract.functions.getRecord(index).call()
    print(f"Record {index}: PatientID={record[0]}, Type={record[1]}, Value={record[2]}, Timestamp={record[3]}")

# --- Run it ---
store_health_data("PAT001", "Heart Rate", "75 BPM")
store_health_data("PAT002", "SpO2", "98%")
store_health_data("PAT003", "Body Temp", "36.8 C")

total = contract.functions.getTotalRecords().call()
print(f"Total records on blockchain: {total}")

for i in range(total):
    get_health_data(i)
```
